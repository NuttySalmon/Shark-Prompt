```python
import pandas as pd
```

```python
df = pd.read_csv("raw.txt",header=None, sep='\t')
df
```

Extracts types to individual columns

```python
df.drop([1,2], axis=1, inplace=True)
df['types'] = df[3].map(lambda x: x[1:-1].split(', '))  # get types
df = df.join(df.types.str.join('|').str.get_dummies().astype(bool))
df.drop([3, 'types'], axis=1, inplace=True)
df
```

Filter by types to get nouns

```python
excluded = ['conjunction', 'predeterminer', 'auxiliary verb', 'indefinite article']
for word_type in excluded:
    df = df[df[word_type] == False]
```

```python
df = df[df.noun]
```

```python
df
```

Write to file

```python
df[0].sort_values().to_csv('1500.txt', header=False, index=False)
```
